```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def main():
    plt.figure(figsize=(25, 20))
    
    df = pd.read_csv("./conway_algorithm_results/win_prob_chart_2_2.csv", header=0, index_col=0)
    
    sns.set(font_scale=4)
    plot = sns.heatmap(data=df, annot=True, fmt=".3f", linewidths=3, linecolor="black", cmap="Spectral_r")
    
    plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
    plot.set_yticklabels(plot.get_yticklabels(), rotation=0)
    
    plot.xaxis.tick_top()
    
    plt.show()
```

```python
if __name__ == '__main__':
    main()
```
